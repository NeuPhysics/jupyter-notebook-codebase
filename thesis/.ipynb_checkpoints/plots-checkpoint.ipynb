```python


constructive_edp = 500



constructive1_predictions = np.transpose( np.genfromtxt( 'rabi/constructive1_k1_0.95_k2_2.6_a1_0.025_a2_0.4-predictions.csv', delimiter=",") )

constructive1 = np.transpose( np.genfromtxt( 'rabi/constructive1_k1_0.95_k2_2.6_a1_0.025_a2_0.4.csv', delimiter=",") )
constructive1_th = 1.


constructive2 = np.transpose( np.genfromtxt( 'rabi/constructive2_k1_0.95_k2_0_a1_0.025_a2_0.csv', delimiter=",") )
constructive2_th = 0.2

constructive3 = np.transpose( np.genfromtxt( 'rabi/constructive3_k1_2.6_k2_0_a1_0.4_a2_0.csv', delimiter=",") )
constructive3_th = 0.0588235


# constructive2=
```

```python
c_plt, c_ax = plt.subplots()

# plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

sfps = 10

plt.axhline(y=constructive1_th, linewidth=1, color='r')
plt.axhline(y=constructive2_th, linewidth=1, color='b')
plt.axhline(y=constructive3_th, linewidth=1, color='k')

c_plt = plt.plot(constructive1[0][0::sfps],constructive1[1][0::sfps],'r-',\
                              constructive2[0][0::sfps],constructive2[1][0::sfps],'b--',\
                              constructive3[0][0::sfps],constructive3[1][0::sfps],'k-.',\
                              constructive1_predictions[0][0::sfps], constructive1_predictions[1][0::sfps], 'r:',\
                              markersize=ftsz*0.4, linewidth=2, fillstyle='full', markeredgecolor='red', markeredgewidth=0.0)
plt.xlabel('Normalized Distance $\omega_{\mathrm{m}}r$', fontsize=ftsz)
plt.ylabel('Probability', fontsize=ftsz)
plt.ylim([0,1.1])
plt.xlim([0,constructive_edp/2])
#plt.xlim([0,375000])
# plt.legend(['Rabi Formula','Numerical Solution'], fontsize=ftsz,frameon=False,loc=1)

c_ax.tick_params('both', length=12, width=2, which='major', labelsize=ftsz, top='off', right='off')
c_ax.tick_params('both', length=8, width=1, which='minor', labelsize=ftsz, top='off', right='off')

c_majorLocator = MultipleLocator(50000)
c_majorFormatter = FormatStrFormatter('%d')
c_minorLocator = MultipleLocator(10000)

c_majorLocatory = MultipleLocator(0.2)
c_majorFormattery = FormatStrFormatter('%0.1f')
c_minorLocatory = MultipleLocator(0.04)


c_ax.xaxis.set_major_locator(c_majorLocator)
c_ax.xaxis.set_major_formatter(c_majorFormatter)

c_ax.yaxis.set_major_locator(c_majorLocatory)
c_ax.yaxis.set_major_formatter(c_majorFormattery)

# xlabels = [item.get_text() for item in interferenceax.get_xticklabels()]
c_xlabels = ['', '0','0.5','1.0','1.5','2.0', '2.5', '3.0', '3.5', '4.0']
c_ax.set_xticklabels(c_xlabels)

# for the minor ticks, use no labels; default NullFormatter
c_ax.xaxis.set_minor_locator(c_minorLocator)
c_ax.yaxis.set_minor_locator(c_minorLocatory)

plt.savefig('rabiOscillationsNeutrinoCoincidence-two-frequencies-constructive.pdf',bbox_inches='tight')
```
