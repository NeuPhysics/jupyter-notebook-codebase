## Halo Problem

```python
hp_ctheta = np.linspace(-1,1)
hp_coeff = 1 - hp_ctheta
```

```python
plt.plot( hp_ctheta, hp_coeff )
```

```python
hp_cosh_z = np.linspace(0,10,10000)
hp_cosh_cosh = np.cosh( 2*(hp_cosh_z-5) )
```

```python
plt.semilogy( hp_cosh_z, hp_cosh_cosh )
# plt.title("$\cosh (2 (5- x))$")
plt.xlabel("x")
plt.ylabel("$\cosh (2 (z-5))$")

plt.savefig("cosh.pdf", bbox_inches='tight')
```
