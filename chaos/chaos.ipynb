```python
import numpy as np
```

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
r = 0.2
a = 10.0
b = 10.0
g = 1.0
```

```python
def sys(x,t):
    dx0dt = - r * x[2] * (x[2]**2 - a**2) -  g * x[2] * x[3]**2
    dx1dt = -r * x[3] * (x[3]**2 - b**2) - g * x[3] * x[2]**2
    dx2dt = x[0]
    dx3dt = x[1]
    
    dxdt = [dx0dt,dx1dt,dx2dt,dx3dt]
    
    return dxdt
```

```python
init = [10,0,5,5]
t = np.linspace(0,200,200000)
t_fine = np.linspace(0,200,2000000)
```

```python
x = odeint(sys,init,t)
x_fine = odeint(sys,init,t_fine)
```

```python
plt.figure(figsize=(10,10))

plt.plot(x[:,2],x[:,3],'k-')
plt.plot(x_fine[:,2],x_fine[:,3],'r--')

plt.show()
```
